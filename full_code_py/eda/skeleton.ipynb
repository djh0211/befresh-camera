```python
import pandas as pd
from pathlib import Path
import pickle
import numpy as np
from matplotlib import pyplot as plt

file_path = './time_history.pickle'
data = None
if Path(file_path).is_file():
    with open(file_path, 'rb') as fr:
        data = pickle.load(fr)
        

```

# Connection Success Percentage

```python
from collections import Counter

counter = Counter(list(map(lambda x:x[0], data)))
dic = dict(counter)

labels = 'Success', 'Fail'
sizes = []
for k in labels:
    if k=='Success':
        sizes.append(dic.get(True) if True in dic else 0)
    else:
        sizes.append(dic.get(False) if False in dic else 0)
colors = ['blue', 'red']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')

plt.show()
```

```python
time_data = list(map(lambda x:x[1], data))
grouped_numbers = [int(i) for i in time_data]
counter = Counter(grouped_numbers)
groups = list(counter.keys())
counts = list(counter.values())

plt.figure(figsize=(10,6))
plt.bar(groups, counts)
plt.xlabel('Time Group')
plt.ylabel('Count')
plt.title('Connection + Data Transfer Time Plot')
plt.xticks(rotation=45)
plt.text(25, 30, f'max elapsed : {max(time_data)}')
plt.show()
```
